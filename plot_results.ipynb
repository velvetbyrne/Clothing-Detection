```python
# Import pandas
import pandas as pd
import matplotlib.pyplot as plt 

# reading csv file 
df = pd.read_csv("results.csv")
print(df.head())
```

```python
df.columns
```

```python
 
ax = plt.gca() 


# use plot() method on the dataframe 
df.plot( x= '                  epoch', y= '           val/box_loss',ylabel='metric', ax = ax, ylim=(0, 6) ) 
df.plot( x= '                  epoch', y= '           val/cls_loss',ylabel='metric', ax = ax, ylim=(0, 6)  ) 
fig = df.plot( x= '                  epoch', y= '           val/dfl_loss',ylabel='losses', ax = ax, ylim=(0, 6)  ).get_figure()

# Alternatively, you can use .set_index 
# to set the data of each axis as follows: 
# df_days_calories.set_index('day')['calories'].plot(); 
fig.savefig(dpi=300,fname='fig_validation.png')
```

```python
 
ax = plt.gca() 


# use plot() method on the dataframe 
df.plot( x= '                  epoch', y= '       metrics/mAP50(B)',ylabel='metric', ax = ax ) 
fig = df.plot( x= '                  epoch', y= '    metrics/mAP50-95(B)',ylabel='mean Average Precision', ax = ax ).get_figure() 
# df.plot( x= '                  epoch', y= '         train/dfl_loss',ylabel='unit', ax = ax ) 

# Alternatively, you can use .set_index 
# to set the data of each axis as follows: 
# df_days_calories.set_index('day')['calories'].plot(); 
fig.savefig(dpi=300,fname='fig_mAP.png')
```

```python
 
ax = plt.gca() 


# use plot() method on the dataframe 
df.plot( x= '                  epoch', y= '         train/box_loss',ylabel='unit', ax = ax ) 
df.plot( x= '                  epoch', y= '         train/cls_loss',ylabel='unit', ax = ax ) 
fig = df.plot( x= '                  epoch', y= '         train/dfl_loss',ylabel='loss', ax = ax ).get_figure() 

# Alternatively, you can use .set_index 
# to set the data of each axis as follows: 
# df_days_calories.set_index('day')['calories'].plot(); 
fig.savefig(dpi=300,fname='fig_training.png')
```
